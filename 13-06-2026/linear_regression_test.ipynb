```python
import joblib
import pandas as pd
```

```python
model_filename = 'artefacts/final_sgd_reg_model.joblib'
preprocessor_filename = 'artefacts/data_preprocessor.joblib'
```

```python
loaded_model = joblib.load(model_filename)
loaded_preprocessor = joblib.load(preprocessor_filename)
```

```python
loaded_model, loaded_preprocessor
```

```python
FEATURE_COLUMNS = ["crim", "zn", "indus", "chas", "nox", "rm", "age", "dis", "rad", "tax", "ptratio", "black", "lstat"]
```

```python
def make_predictions_from_list(input_list, feature_names, loaded_model, loaded_preprocessor):
    # Create a DataFrame from the input list
    x_input = pd.DataFrame([input_list], columns=feature_names)
    
    # Preprocess the input data
    x_input_processed = loaded_preprocessor.transform(x_input)
    
    # Make a prediction using the loaded model
    prediction = loaded_model.predict(x_input_processed)
    
    return prediction[0]
```

```python
test_feature_list = [
    50,   # crim
    0,    # zn
    18.1, # indus
    0,    # chas
    0.7,  # nox
    4.0,  # rm
    100,  # age
    1.0,  # dis
    24,   # rad
    666,  # tax
    20.2, # ptratio
    300,  # black
    30    # lstat
]
```

```python
predicted_value = make_predictions_from_list(test_feature_list, FEATURE_COLUMNS, loaded_model, loaded_preprocessor)
```

```python
print(f"Predicted value: {predicted_value:.2f}")
```
